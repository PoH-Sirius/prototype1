```python
import numpy as np

def create_translation_matrix(tx, ty, tz):
  return np.array([
      [1, 0, 0, tx],
      [0, 1, 0, ty],
      [0, 0, 1, tz],
      [0, 0, 0, 1]
  ])

def create_rotation_matrix_y(theta):
  c = np.cos(theta)
  s = np.sin(theta)

  return np.array([
      [c, 0, s, 0],
      [0, 1, 0, 0],
      [-s, 0, c, 0],
      [0, 0, 0, 1]
  ])
```

```python
hand_point = np.array([1.0, 0.5, 0.0, 1.0])

angle = np.radians(45)
R = create_rotation_matrix_y(angle)
T = create_translation_matrix(2.0, 0.0, -1.0)

M = T @ R

virtual_hand_point = M @ hand_point
```

```python
print(f"現実の座標: {hand_point[:3]}")
print(f"変換行列 M:\n{M}")
print(f"仮想空間の座標: {virtual_hand_point[:3]}")
```
